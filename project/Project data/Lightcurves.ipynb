```python
# Read in lightcurves from the lightcurve_generator.py and Plasticc_data.ipynb
# plot  example lightcurve 
import pandas as pd
import pickle
import matplotlib.pyplot as plt
```

```python
def plot_timeseries(df: pd.DataFrame, indx=0 ) -> None:
    '''Plot the generated time series data.
        Args:
            df: Pandas DataFrame with the generated data
            indx: index of the time series to plot
        '''
    df[f'y{indx}' ].plot(style='o')  

    plt.show()
```

```python
# read sinusioid lightcurves - single frequency
df_single = pd.read_feather("timeseries_single_frequency.feather")
plot_timeseries(df_single, indx=50)
```

```python
# read sinusioid lightcurves - muliple frequencies
df_multiple = pd.read_feather("timeseries_multiple_frequency.feather")
plot_timeseries(df_multiple, indx=50)
```

```python
# read sinusioid lightcurves - muliple frequencies sparsely sampled light curve
df_sampled = pd.read_feather("timeseries_multiple_frequency_sampled.feather")
plot_timeseries(df_sampled, indx=10)
```

```python
# Read PLASTICC lightcurves
plasticc_metadata = pd.read_pickle('plasticc_lightcurves_g_metadata.pkl')
with open('plasticc_lightcurves_g.pkl', 'rb') as file:
    plasticc_LC = pickle.load(file)

plasticc_LC[plasticc_metadata['object_id'].iloc[20]].plot(style='o')
```
